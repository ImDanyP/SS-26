# Anatomía de una Gráfica

Primero es necesario, conocer la "anatomía" de una gráfica generada por la libreria, para asi saber que se busca modificar/agregar.

Estas son las caracteristicas generales de una gráfica:

![Anatomia de una gráfica](resources/anatomy.webp)

Cada elemento puede modificar de la siguiente forma:



```python
x = np.linspace(-2, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(
    x,
    y,
    color='red',
    linewidth=3,
    linestyle='--', # '-', '--', '-.', ':'
    marker='^',# 'o', 's', '^', 'v', '*', '+', 'x', 'D'
    markersize=5,
    markerfacecolor='yellow',
    markeredgecolor='black',
    label='sin(x)'
)

ax.scatter(
    x,
    y,
    s=50, #tamaño del marcador
    c='purple',
    marker='s',
    alpha=0.3 #transparencia
)

ax.set_title(
    "Título",
    fontsize=20,
    color='darkblue',
    fontweight='bold',
    pad=20
)
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")

ax.set_xticks([-2,0,2,4,6,8,10]) #se pueden usar np.arange(0,11,2) para generar los mismos valores
ax.set_yticks([-1,-0.5,0,0.5,1]) # o se puede usar np.linspace(-1,1,5)

ax.legend(
    loc='upper right', # 'best', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
    fontsize=12,
    frameon=True,
    shadow=True,
    title='Funciones'
)

ax.grid(
    True,
    color='gray',
    linestyle='--', # '-', '--', '-.', ':'
    linewidth=0.8,
    alpha=0.5 #transparencia
)

plt.show()
```

# Anotaciones

Ademas de los detalles que se pueden agregar a los elementos de una gráfica es posible hacer anotaciones sobre la misma de la siguiente forma:

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('Máximo local', #texto a anotar
            xy=(2, 1),#coordenadas del punto a anotar
            xytext=(3, 1.5), #coordenadas del texto
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```

Otros ejemplos:

```python
fig, ax = plt.subplots(figsize=(3, 3))
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]
annotations = ["A", "B", "C", "D", "E"]
ax.scatter(x, y, s=20)

for xi, yi, text in zip(x, y, annotations):
    ax.annotate(text,
                xy=(xi, yi), xycoords='data',
                xytext=(1.5, 1.5), textcoords='offset points')
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

ann = ax.annotate("Test",
                  xy=(0.2, 0.2), xycoords='data',
                  xytext=(0.8, 0.8), textcoords='data',
                  size=20, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="-|>",
                                  connectionstyle="arc3,rad=-0.2",
                                  fc="w"))
```

# Regiones de la gráfica

El manejo de los bordes en una gráfica es importante, por default la libreria intentara que en la gráfica se muestre todo el conjunto de datos graficado, si nuestro conjunto de datos no esta pensado para eso esto puede ser inconveniente esto puede ser modificado de la siguiente manera:

```python
# Consideremos una función sen(x)/x 

x = np.linspace(-10, 10, 200) #La función solo esta definida en el intervalo [-10,10]
y = np.sinc(x)

fig, ax = plt.subplots()
ax.plot(x, y)
```

Moviendo los margenes:

```python
x = np.linspace(-10, 10, 200) 
y = np.sinc(x)


fig, ax = plt.subplots()
ax.plot(x, y)
plt.xlim(-5, 5)
plt.ylim(-0.5, 1)
```

```python
x = np.linspace(-10, 10, 200) 
y = np.sinc(x)


fig, ax = plt.subplots()
ax.plot(x, y)
plt.xlim(-10, -2.5)
plt.ylim(-0.5, 1)
```

# Resolución de Imágenes

La libreria tambien nos ofrece la opción de guardar directamente las gráficas obtenidas en la carpeta de ejecución del código, pero el tamaño de estas suele ser en un formato pequeño, que si bien es legible (con una buena calidad) en el monitor, esta resolución suele ser insuficiente para otros formatos, como posters, presentaciones, papers, tesis, etc. 

En este caso hay dos soluciones, hacer que la imagen sea de un tamaño mayor (para asi aumentar artificialmente su resolución) o guardarlo en un formato vectorial (como pdf o svg).

### Aumentando el tamaño

```python
x = np.linspace(-10, 10, 200)
y = np.sinc(x)
 

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, y)
```

O de otra forma:

```python
x = np.linspace(-10, 10, 200) 
y = np.sinc(x)


fig, ax = plt.subplots(figsize=(33.1, 46.8))
ax.plot(x, y)
```

Finalmente para guardar el archivo usamos la linea de código:

(En caso de usar un cuaderno de Jupyter es posible guardar la imagen directamente al pasar el mouse por la esquina superior derecha de la imagen y haciendo click en el ícono del floppy disk (disquete))

```python
x = np.linspace(-10, 10, 200) 
y = np.sinc(x)


fig, ax = plt.subplots(figsize=(33.1, 46.8))
ax.plot(x, y)
plt.savefig("resultados/grafico.png", dpi=300) #dpi es la resolución del gráfico, entre más alto el dpi, mejor calidad tendrá el gráfico.
```

Comparando el resultado:

![Imagen guardada](resultados/grafico.png)

### Usando un formato vectorial

Los formatos vectoriales como .pdf tienen la gran ventaja de no perder resolución sin importar cuanto zoom se haga o cuanto se amplíe la imagen

```python
x = np.linspace(-10, 10, 200) 
y = np.sinc(x)


fig, ax = plt.subplots()
ax.plot(x, y)
plt.savefig("resultados/grafico.pdf",#Se puede guardar el gráfico en formato pdf, svg, eps, etc. 
            bbox_inches='tight' #Ajusta el tamaño del gráfico al contenido, eliminando los márgenes en blanco.
            ) 
```
